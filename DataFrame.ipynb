```python
# Repetição com while


dados = {
    "Nome": [ 'ana', 'bruno', 'carlos'],
    "Idade": [25,30,22],
    "Cidade": ['SP', 'RJ','BH']
}

df = pd.DataFrame(dados)
indice = 0
encontrou = False

while indice < len(df) and not encontrou:
  if df.loc[indice, 'Nome'] == 'bruno':
    print(f"Ana encontradada indice {indice}")
    encontrou = True
    indice += 1
```

```python
#função para aplicar desconto

dados = {
    "Produto": ['caneta', 'caderno', 'livro'],
    'Preço': [5,30,45],
}

df = pd.DataFrame(dados)


def aplica_desconto(preco):
  desconto = preco * 0.9
  return desconto

df['Preço com desconto'] = df['Preço'].apply(aplica_desconto)
df
```

```python
#1 Crie uma função para calcuular o desconto de PCD para os clientes, mostrando o valor do carro e o valor do desconto
#. Sabendo que o valor de desconto é de 20% para tdoos os carros



dados = {
    "Cliente": ['A', 'B', 'C', 'D'],
    "Preço": [ 45000, 30000, 55500, 98000],
}

#Cliente, Preço, Desconto, Preço com desconto

df = pd.DataFrame(dados)

def aplicar_desconto(preco):
  desconto = preco * 0.8
  return desconto

df['Preço com Desconto']= df['Preço'].apply(aplicar_desconto)
df['Desconto']= df['Preço']- df['Preço com Desconto']
df
```

```python
def classificar_rotacao(rotacao):
  if rotacao < 900:
    return 'Muito baixo'
  elif 900< rotacao <= 1100:
    return 'Baixo'
  elif 1100 < rotacao < 1400:
    return 'Médio'
  elif 1400 < rotacao < 1500:
    return 'Alto'
  else:
    return 'Muito alto'

df ['Categoria Rotação']= df['Velocidade Rotacao [rpm]'].apply(classificar_rotacao)
df
```
