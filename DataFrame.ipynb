```python
#1.Crie o seguinte dataframe:
    
   # Nome: ana, laura, martim, julia, marcelo
  #  Salario: 2500, 2800, 3200, 2400, 3900
 #   desconto: 10,12,10,15,18
    
#Calcule o desconto e mostre o valor  que cada pessoa irá receber após o 
#decsonto

#Salve este arquivo em csv e volte a ler o arquivo que foi salvo
import pandas as pd

dados2 = { 'Nome': ['ana', 'laura', 'martim', 'julia', 'marcelo'],
          'Salario': [2500, 2800, 3200, 2400, 3900],
          'Desconto': [35,31,10,30,54]
}


df= pd.DataFrame(dados2)


df['Valor_Descontado']= df['Salario'] * (df['Desconto']/100)


df['Valor_Receber']= df['Salario'] - df['Valor_Descontado']

print(df['Valor_Receber'].min())
print(df['Valor_Receber'].max())
print(df['Valor_Receber'].mean())
print(df['Valor_Receber'].median())
```

```python
dados = {
    'Nome': ['ana', 'laura', 'martim'],
    'Peso': [70,80,90],
    'Altura': [1.65,1.75,1.80]
}

df = pd.DataFrame(dados)
#Criar uma nova coluna IMC

for index, row in df.iterrows():
    imc = row['Peso']/ row['Altura']**2
    df.at[index, 'IMC']= imc

df
```

```python
dados5 = { 'Name': ['ana', 'laura', 'martim'],
          'Idade': [21,23,35],
          'Salario': [2500, 2800, 3200],
          'Cidade': ['RJ', 'SP', 'MG']
}

df= pd.DataFrame(dados5)

# Adicionar nova estrutura de classficação baseada na idade 

for index, row in df.iterrows():
    if row['Idade']<30:
        df.at[index, 'Classificação']= 'Jovem'
    elif row['Idade']<50:
        df.at[index, 'Classificação']= 'Adulto'
    else:
        df.at[index, 'Classificação']= 'Idoso'
print(df)
```

```python
#Para o exercício anterior, crie uma nova coluna chamada Metropole, 
# se a Cidade for São Paulo ou Rio de Janeiro,
#escreva sim, caso contrário, escreva não.

dados6 = { 'Name': ['ana', 'laura', 'martim'],
          'Idade': [21,23,35],
          'Salario': [2500, 2800, 3200],
          'Cidade': ['RJ', 'SP', 'MG'],
          'Metropole':[]
}

df= pd.DataFrame(dados5)

# Adicionar nova estrutura de classficação baseada na idade 

for index, row in df.iterrows():
    if row['Idade']<30:
        df.at[index, 'Classificação']= 'Jovem'
    elif row['Idade']<50:
        df.at[index, 'Classificação']= 'Adulto'
    else:
        df.at[index, 'Classificação']= 'Idoso'


for index, row in df.iterrows():
    if row['Cidade']== 'RJ' or row['Cidade']=='SP':
        df.at[index, 'Metropole']= 'Sim'
    else:
        df.at[index, 'Metropole']= 'Não'

df
```
