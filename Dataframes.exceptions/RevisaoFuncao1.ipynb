```python

#função dentro de função


def eleva_quadrado():
    a = float(input("Digite altura"))
    b = float(input("Digite a base"))
    quadrado = a**2 + b**2
    return quadrado
quadrado = eleva_quadrado()
def hipotenusa_cal():
    h = math.sqrt(quadrado)
    return h
hipotenusa = hipotenusa_cal()
print(quadrado)
print(hipotenusa)

```
